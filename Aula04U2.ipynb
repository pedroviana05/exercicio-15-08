```python
import matplotlib.pyplot as plt

#dados
x = [1, 2, 3, 4, 5]
y = [2, 4, 1, 3, 5]

#criar um grafico de linha
plt.bar(x,y)

#adicionar rótulos aos eixos
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')

#adicionar um titulo ao gráfico
plt.title('Exemplo de Grafico de Linha')

#mostrar o grarfico
plt.show()





```

```python
import matplotlib.pyplot as plt
#dados de exemplo
meses = ['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio']
vendas = [2500, 2000, 1800, 1200, 3400]
# Criar um grafico de barrar
plt.bar(meses, vendas, color ='green')
#adiciomar rotulos aos eixos
plt.xlabel('\nMês')
plt.ylabel('Vendas por mês')

#adicionar um titulo ao grafico
plt.title('\nVendas Mensais')

#mostrar o grafico
plt.show()
```
